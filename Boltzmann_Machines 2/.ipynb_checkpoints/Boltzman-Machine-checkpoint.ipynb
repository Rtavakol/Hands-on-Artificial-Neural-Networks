```python
# Importing the libraries
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.parallel
import torch.optim as optim
import torch.utils.data
from torch.autograd import Variable
```

```python
movies = pd.read_csv('ml-1m/movies.dat', sep='::', header=None, engine='python', encoding='latin-1')
users = pd.read_csv('ml-1m/users.dat', sep='::', header=None, engine='python', encoding='latin-1')
ratings =pd.read_csv('ml-1m/ratings.dat', sep='::', header=None, engine='python', encoding='latin-1')
```

```python
#Preparing the training and test sets
training_set = pd.read_csv('ml-100k/u1.base', delimiter= '\t')
training_set = np.array(training_set, dtype = 'int')
test_set = pd.read_csv('ml-100k/u1.test', delimiter= '\t')
test_set = np.array(test_set, dtype = 'int')
```

```python
# Getting the number of users and movies
nb_users = int(max(max(training_set[:, 0]), max(test_set[:, 0])))
nb_movies = int(max(max(training_set[:,1]), max(test_set[:,1])))
```

```python
# Conver data in array with users in lines and movies in columns
###NOTE we will not make a matrix we make list of list because of torch input requirement###
def convert(data):
    new_data = []
    for id_users in range(1, nb_users + 1):
        id_movies = data[:, 1][data[:, 0] == id_users]
        id_ratings = data[:, 2][data[:, 0] == id_users]
        ratings = np.zeros(nb_movies)
        ratings[id_movies - 1] = id_ratings  # id_movies - 1 because id starts from 1 but list position index starts from 0
        new_data.append(list(ratings))
    return new_data
```

```python
training_set_matrix = convert(training_set) # in fact list of list, NOT matrix
test_set_matrix = convert(test_set)
print(len(training_set_matrix), len(test_set_matrix))
```

```python
# Converting the data into Torch tensors
training_set_matrix_tensor = torch.FloatTensor(training_set_matrix)
test_set_matrix_tensor = torch.FloatTensor(test_set_matrix)
```

```python
# Convert ratings into binary
training_set_matrix_tensor[training_set_matrix_tensor == 0] = -1
training_set_matrix_tensor[training_set_matrix_tensor == 1] = 0
training_set_matrix_tensor[training_set_matrix_tensor == 2] = 0
training_set_matrix_tensor[training_set_matrix_tensor >= 3] = 1
```

```python
test_set_matrix_tensor[test_set_matrix_tensor == 0] = -1
test_set_matrix_tensor[test_set_matrix_tensor == 1] = -1
test_set_matrix_tensor[test_set_matrix_tensor == 2] = -1
test_set_matrix_tensor[test_set_matrix_tensor >= 3] = 1
```

```python
# Creating Architecture of Neural Network
class RBM():
    def __init__(self, nv, nh): # nv: number of visible nodes, nh: number of hidden node
        self.W = torch.randn(nh, nv) # W is weight, Returns a tensor filled with random numbers 
                                        # from a normal distribution
        
        
        self.a = torch.randn(1, nh) # Initialize bias of hidden nodes
        self.b = torch.randn(1, nv) # Initialize bias of visible nodes
    
    def sample_h(self, x): # sampling the hidden node
        wx = torch.mm(x, self.W.t())  
        activation = wx + self.a.expand_as(wx) # expand to add bias to each line of wx
        p_h_given_v = torch.sigmoid(activation)# Probability of h given v
        return p_h_given_v, torch.bernoulli(p_h_given_v)
    
    def sample_v(self, y): #sampling the vissible node
        wy = torch.mm(y, self.W)
        activation = wy + self.b.expand_as(wy)
        p_v_given_h = torch.sigmoid(activation)
        return p_v_given_h, torch.bernoulli(p_v_given_h)
    
    def train(self, v0, vk, ph0, phk):
        """
        v0: input vector of ratings
        vk: visible node after k sampling
        ph0: vector of probability of hidden node at 1st sampling given visible node v0
        phk: vector of probability of hidden node at kth sampling given visible node vk
        """
        
        
        # we are implementing code lines of 8 - 10 on page 28th of the paper that is in the current folder
        # paper is "An Introduction to Restricted Boltzmann Machines"
        
        self.W += (torch.mm(v0.t(), ph0) - torch.mm(vk.t(), phk)).t()
        self.b += torch.sum((v0 - vk), 0)
        self.a += torch.sum((ph0 - phk), 0)
```

```python
nv = len(training_set_matrix_tensor[0])
nh = 100 # hidden nodes are features that we are going to detect, just use a relevant number, in this case 
            # is the number of actors, awards, actresses, ... 
batch_size = 100
rbm = RBM(nv, nh)
```

Training the RBM

```python
nb_epoch = 10
for epoch in range(1, nb_epoch + 1):
    train_loss = 0
    s = 0.
    for id_user in range(0, nb_users - batch_size, batch_size):
        vk = training_set_matrix_tensor[id_user : id_user + batch_size]
        v0 = training_set_matrix_tensor[id_user : id_user + batch_size]
        ph0,_ = rbm.sample_h(v0)
        for k in range(10):  # Contrastive Divergence loop
            _,hk = rbm.sample_h(vk)
            _,vk = rbm.sample_v(hk)
            vk[v0<0] = v0[v0<0] # those movies that are not rated
        # Before applying train functiontrain(v0, vk, ph0, phk) we need to find phk
        phk,_ = rbm.sample_h(vk)
        rbm.train(v0, vk, ph0, phk)
        
        train_loss += torch.mean(torch.abs(v0[v0>=0] - vk[v0>=0]))
        s += 1.
    print('epoch: ' + str(epoch) + ' loss: ' + str(train_loss/s))
```

Testing the RBM

```python
test_loss = 0
s = 0.
for id_user in range(nb_users):
    v = training_set_matrix_tensor[id_user : id_user + 1]
    vt = test_set_matrix_tensor[id_user : id_user + 1]
    if len(vt[vt>=0]) > 0:
        _,h = rbm.sample_h(v)
        _,v = rbm.sample_v(h)
        test_loss += torch.mean(torch.abs(vt[vt>=0] - v[vt>=0]))
        s += 1.
print('test_loss: ' + str(test_loss/s))
```
